## Appling KNN for Different values of K

```
def fun(dist,k):
  knn = []
  for i in range(0,k):
    knn.append(dist[i])
  
  # print(knn)
  knn_label = []
  for i in range(0,len(knn)):
    knn_label.append(knn[i][1])
  
  knn_value = Counter(knn_label)
  res = knn_value.most_common(1)[0][0]

  # print(res)
  return res

def compare(y_predict,y_true):
    print(accuracy_score(y_true,y_predict))


all_dist = []
euclid_dist = []
model_label = []

for k in range(1,10):
  for validation_data_row in validation_data:
    i=0
    for train_data_row in train_data:
      # print(train_data_without_label)
      row_diff = np.power((train_data_row - validation_data_row),2)
      # print(row_diff)
      row_euclid_diff = int(np.sqrt(np.sum(row_diff)))
      # print(row_euclid_diff)
      x = train_data_label[i][0]
      # print(x)
      euclid_dist.append((row_euclid_diff,x))
      i=i+1
    # print(euclid_dist)
    
    dist2 = sorted(euclid_dist, key=lambda x: x[0])
    model_label.append(fun(dist2,k))
    euclid_dist.clear()

  validation_label = []
  for i in range(0,len(validation_data_label)):
    validation_label.append(validation_data_label[i])
    # print(validation_data_label[i])

  print("Accuracy Score")
  print('k ==',k)
  compare(model_label,validation_label)
  model_label.clear();
  validation_label.clear()
```
